```python
from bs4 import BeautifulSoup
from requests_html import HTMLSession
```

```python
sections_urls = {
    'sport_news':'https://www.youm7.com/Section/%D8%A3%D8%AE%D8%A8%D8%A7%D8%B1-%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6%D8%A9/298/1',
    'politics_news' : 'https://www.youm7.com/Section/%D8%B3%D9%8A%D8%A7%D8%B3%D8%A9/319/1'
}
```

```python
def get_section_news_links(section_url):
    news_links = []
    news_page = HTMLSession().get(section_url)
    news_soup = BeautifulSoup(news_page.content,'lxml')
    news_section = news_soup.find('div',class_='section-news')
    all_news = news_section.find_all('div',class_='bigOneSec')
    for news in all_news:
        base_url = 'https://www.youm7.com'
        url = news.find('div').find('a')['href']
        news_links.append(f"{base_url}{url}")
    return news_links
```

```python
get_section_news_links(sections_urls['sport_news'])
```

```python
def get_article_data(article_link):
    data = {}
    page = HTMLSession().get(article_link)
    soup = BeautifulSoup(page.content,'lxml')
    article_content = soup.find('div',class_='article-cont')
    title = article_content.find('div',class_='articleHeader').find('h1').text.strip()
    date = article_content.find('div',class_='articleHeader').find('span').text.strip()
    article_body = article_content.find('div',class_='articleCont').find('div',id='articleBody')
    article_paraghraphs = article_body.find_all('p')
    article = """"""
    for paragraph in article_paraghraphs:
        article+=paragraph.text.strip()
    data['Title'] = title
    data['Date'] = date
    data['Article'] = article
    return data
```

```python
def get_section_articles(section_urls):
    articles = []
    for url in section_urls:
        articles.append(get_article_data(url))
    return articles
```

```python
sports_links = get_section_news_links(sections_urls['sport_news'])
```

```python
get_section_articles(sports_links)
```

```python
politics_links = get_section_news_links(sections_urls['politics_news'])
```

```python
get_section_articles(politics_links)
```
